```python
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv, find_dotenv
import os

def fetch_and_parse_webpage():
    # Load environment variables from .env file
    dotenv_path = find_dotenv()
    if not dotenv_path:
        raise ValueError("Could not find .env file")

    load_dotenv(dotenv_path)
    print(f"Loaded environment from: {dotenv_path}")

    # Retrieve the URL from environment variables
    url = os.getenv('webpage_url')

    # Debugging: Print the loaded URL
    print(f"URL from environment variable: {url}")

    # Check if the environment variable is loaded correctly
    if not url:
        raise ValueError("Missing environment variable: 'webpage_url'")

    # Make a request to the website
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup and lxml
        soup = BeautifulSoup(response.text, 'lxml')
        
        # Print the prettified HTML content
        print(soup.prettify())
        
        # Example: Extract and print all quotes
        quotes = soup.find_all('span', class_='text')
        for quote in quotes:
            print(quote.get_text())

    else:
        print(f"Failed to retrieve the webpage: {response.status_code}")

# Example usage
try:
    fetch_and_parse_webpage()
except ValueError as e:
    print(e)
except requests.exceptions.RequestException as e:
    # Catch any request-related errors
    print(f"Request failed: {e}")
```

## Working with Multiple Pages and Items from BooksToScrape.com

```python
# Target: get the Title of every book with a 2 stars rating using toscrape.com
```

```python
# save the URL in a variable, get the number of pages to be changable using {} to be able to scrape from multiple pages later
base_url = 'http://books.toscrape.com/catalogue/category/books_1/page-{}.html'
```

```python
page_num = 12
base_url.format(page_num)
```

```python
res = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
products = soup.select(".product_pod")
```

```python
type(products)
```

```python
# check if it works for just 1 example now, for later
ex = products[0]
```

```python
# we can see the important things for us here is: class and title
ex
```

```python
# let's see if this ex is rated as 2 stars or not?

# there many ways to check that for instance using string
# convert the ex to a string then check if the class =  'star-rating Tow' ? and return True or False

'star-rating Two' in str(ex)
```

```python
# another way (more official tactic) using beautifulsoup, checking is this class 'star-rating Tow' present
ex.select(".star-rating Three") # will get back an empty list

# pay attention the class name you"re calling has a space, use dot to fill out that space
ex.select(".star-rating.Three")

ex.select(".star-rating.Two") # and that's what we're looking for

[] == ex.select(".star-rating.Two") # now we've got what we need
```

```python
# let's get the title, first get the title's tag
ex.select('a')
```

```python
# notice that the first link for the image, 2nd one for the title
ex.select('a')[1]
```

```python
# now get only the title
ex.select('a')[1]['title']
```

```python
# now combine both of them and get them in a for loop and get the task done !

two_star_titles = [] # list to save the titles which is 2 stars rate

for n in range (1,51): # to go through all the pages of the website which is 50 pages
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)

    soup = bs4.BeautifulSoup(res.text, 'lxml')
    books = soup.select(".product_pod")

    for book in books:
        # using the string way to check it
        # if 'star-rating Two' in str(book):
        if len(book.select('.star-rating.Two')) !=0 :
            book_title = book.select('a')[1]['title']
            two_star_titles.append(book_title)
```
